## Анализ карты Пуанкаре BTC-USDT

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# Загрузка исторических данных BTC-USDT
start='2023-01-01'
end='2024-05-25'
interval='1d'
btc_data = yf.download('BTC-USD', start, end, interval)

# Используем закрытые цены
prices = btc_data['Close'].values

# Построение карты Пуанкаре
plt.figure(figsize=(12, 6))
plt.plot(prices[:-1], prices[1:], 'o', markersize=2)
plt.xlabel('$P(t)$')
plt.ylabel('$P(t+1)$')
plt.title('Карта Пуанкаре для BTC-USD'+' c '+start+' по '+end)
plt.show()
```
